# Selecting single or a collection of lensed images

Here we show how to select lensed images.

After the user has selected a collection of lensed images it can plot their event properties.

As an example, we select among the images with the most microlensing signal and show some of their properties.

```python
import numpy as np
import nav22
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lenses = nav22.DB()
```

```python
lenses.lenses
```

```python
lenses.select_images(Nimages=3)
lenses.sample_events(cc_only=True)
lenses.plot_selected()
```

# Constraining the lensed images
Here we show how to constrain the lensed images to show only their HME properties.

For example, let's select the lensed images that have source redshift < 1.5

```python
lenses = nav22.DB()
lenses.lenses = lenses.lenses[lenses.lenses['ZSRC'] < 1.5]
lenses.lenses
```

Now let's see their caustic crossing properties

```python
lenses.plot_all(cc_only=True)
```

Now let's sample them down to the expected lensed systems within this constrain

```python
lenses.sampleNlenses()
lenses.sampled_lenses
```

```python
print(len(lenses.sampled_lenses.index), 'images from', len(np.unique(lenses.sampled_lenses['LENSID'])), 'lensed systems')
```

Lets select the best 5 images and see their caustic crossing properties

```python
lenses.select_images(random=False, ids=lenses.sampled_lenses['ID'].head(5))
lenses.sample_events(cc_only=True)
lenses.plot_selected()
```
